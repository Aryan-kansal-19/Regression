```python
# importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_palette('husl')
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#Reading Iris Dataset csv 
data = pd.read_csv("DataSets\Iris.csv")
```

```python
# Getting Info on DataTypes of Columns For Preprocessing
data.info()
```

```python
#As Dtypes are all Accurate
#Now we Count the Value of Each Type of Flower Species
data['Species'].value_counts()
```

```python
# Now Dropping Id Column as we dont need it as indexing is already Provided
t = data.drop('Id',axis=1)

# Visualing and Plotting the Data for Further Processing the data
g = sns.pairplot(t, hue = 'Species')

plt.show()
```

```python
# now dropping the both Columns Id and Species For Preparing the dataset for Algorithm   
X = data.drop(['Id','Species'],axis=1)

# Veiwing the top 5 rows of the dataset for reveiwing the changes made 
X.head() 
```

```python
# Only Taking Column Species for Putting in algorithm
y = data['Species']
y.head()
```

```python
# Checking Shapes of both X and y Datasets
print("Shape of 'X' DataFrame : ",X.shape)
print("Shape of 'y' Dataframe : ",y.shape)
```

```python
# Putting Both Datasets X,y in K-Nearest Neighbor Algorithm
# As we does not know the Best Value of Neighbors 
# And The Dataset is very small so we have taken the range of neighbors for best Accuracy result
score=[]
n_list=list(range(2,20))
for n in n_list:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X, y)
    y_pred = knn.predict(X)
    score.append(metrics.accuracy_score(y,y_pred))

# Plotting the Graph
plt.plot(n_list,score)
plt.xlabel("Value for n Neighbors in KNN")
plt.ylabel("Accuracy Score")
plt.title("Accuracy Score for each Value of n in KNN")
plt.grid()
plt.show()
```

```python
# Also Checking the Accuracy Score For Logistic Regression Algorithm for best Results
logreg = LogisticRegression()
logreg.fit(X,y)
y_pred = logreg.predict(X)
print("Logistic Regression Accuracy : ",metrics.accuracy_score(y,y_pred))
```

```python
#Splitting The Data of same dataset in test and train data 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=5)
print("X_train :",X_train.shape)
print("y_train :",y_train.shape)
print("X_test :",X_test.shape)
print("y_test :",y_test.shape)
```

```python
# Getting the Accuracy Score For Testing And Training Dataset We Acquired earlier
n_list=list(range(2,20))
score = []
for n in n_list:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    score.append(metrics.accuracy_score(y_pred,y_test))

plt.plot(n_list,score)
plt.xlabel("No. of n Neighbors on KNN")
plt.ylabel("Accuracy Score")
plt.title("Accuracy Score of N Neighbors using KNN")
plt.grid()
plt.show()
```

```python
# Now Applying Logistic regession Algorithm on Testing and Training Dataset 
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
print("Accuracy Score of Logistic Regression : ",metrics.accuracy_score(y_pred,y_test))
```

```python
# As Accuracy By Using KNN Algorithm is Higher than Logistic Regression Algorithm 
# So we will Use KNN Algorithm for Predicting our Query 
# Taking Neighbors n = 8 as accuracy = 1

knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train,y_train)
knn.predict([[6,2,3,4]])    
```
